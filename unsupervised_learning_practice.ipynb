```python
# 1~45 사이의 난수 생성
from numpy import random
x = random.randint(1,46)
print(x)
```

```python
# 0~1 까지의 임의의 부동 소수점 생성
from numpy import random
x = random.rand()
print(x)
```

```python
# 1~45 사이의 7자리 난수 생성
from numpy import random
x=random.randint(1,46, size=(7))
print(x)
```

```python
# 2차원 배열 사용법
from numpy import random
x = random.randint(46,size=(2,6))
print(x)
```

```python
# 배열 값중 하나를 반환
from numpy import random
x = random.choice([3, 5, 7, 9])
print(x)
```

```python
# 이 값으로 구성된 2차원 배열을 생성
from numpy import random
x = random.choice([3, 5, 7, 9], size=(3, 5))
print(x)
```

```python
# 확률 밀도 함수
from numpy import random
x = random.choice([3, 5, 7, 9], p=[0.1, 0.1, 0.8, 0.0], size=(5))
print(x)
```

```python
# 배열에서 고유한 요소로 만들기
import numpy as np
arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])
x = np.unique(arr)
print(x)
```

```python
# 두 배열의 고유한 값만 합쳐서 만들기
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
newarr = np.union1d(arr1, arr2)
print(newarr)
```

```python
# 두 배열의 교집합 찾기
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
print(arr1)
newarr = np.intersect1d(arr1, arr2, assume_unique=True)
print(newarr)
```

```python
# 두 배열의 차집합 찾기 (첫 번째 배열만)
import numpy as np
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
newarr = np.setdiff1d(set1, set2, assume_unique=True)
print(newarr)
```

```python
# 두 배열의 차집합 찾기 (두 배열 모두)
import numpy as np
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
newarr = np.setxor1d(set1, set2, assume_unique=True)
print(newarr)
```
